### Carregando Covariadas

```python
desocup = pd.read_excel("D:/data_sets_cases/desocupacao_bacen.xlsx",index_col='Data',parse_dates=True)
populacao = pd.read_excel("D:/data_sets_cases/pop_tot_2009_2021.xlsx",index_col='Data',parse_dates=True)
renda_med = pd.read_excel("D:/data_sets_cases/renda_media_bacen.xlsx",index_col='Data',parse_dates=True)
ipca = pd.read_excel("D:/data_sets_cases/ipca_bacen.xlsx",index_col='Data',parse_dates=True)

merged_df_1 = pd.merge(desocup, populacao, on='Data')
merged_df_2 = pd.merge(renda_med,ipca, on='Data')

df_cov= pd.merge(merged_df_1,merged_df_2, on="Data")
```

```python
covariadas= pd.merge(df_cov,df_2016['tv_pp'], on="Data")

#Dummys

covariadas['d_covid'] = covariadas.index.map(lambda x: 1 if x > pd.to_datetime('2020-03-01') else 0)

covariadas
```

# MQO 

```python
covariadas_ln =covariadas

covariadas_ln['ln_pop']=covariadas_ln['pop_tot'].apply(np.log)
covariadas_ln['ln_desocup']=covariadas_ln['taxa_de_desocupacao'].apply(np.log)
covariadas_ln['ln_renda']=covariadas_ln['rendimento_medio'].apply(np.log)
covariadas_ln['ln_preco']=covariadas_ln['tv_pp'].apply(np.log)

endog_ln = df_2016['tv_vol'].apply(np.log)

covariadas_ln = covariadas_ln.drop('taxa_de_desocupacao', axis=1)
covariadas_ln = covariadas_ln.drop('pop_tot', axis=1)
covariadas_ln = covariadas_ln.drop('rendimento_medio', axis=1)
covariadas_ln = covariadas_ln.drop('tv_pp', axis=1)
# covariadas_ln = covariadas_ln.drop('IPCA', axis=1)

#########


for month in range(1, 12):
    covariadas_ln[f'dummy_{month}'] = (covariadas_ln.index.month == month).astype(int)

for month in range(1, 12):
    covariadas_ln[f'dummy_{month}'] = covariadas_ln[f'dummy_{month}'] * 2.72
    
for month in range(1, 12):
    dummy = covariadas_ln[f'dummy_{month}']
    covariadas_ln[f'dummy_{month}'] = np.where(dummy != 0, np.log(dummy), 0)

print(covariadas_ln)

df_ols= pd.merge(covariadas_ln,endog_ln, on="Data")
```

```python
import statsmodels.api as sm

# Suponha que você tenha dados como matrizes numpy ou DataFrames do pandas
X = covariadas_ln # Dados de entrada
y = df_ols['tv_vol'] # Variável de resposta

# Adicione uma coluna de uns para representar o intercepto
X = sm.add_constant(X)

# Ajuste o modelo aos dados
model = sm.OLS(y, X)
results = model.fit()
 
# Mostre um resumo dos resultados
print(results.summary())
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan, acorr_breusch_godfrey

# Obtendo os resíduos do modelo
residuals = results.resid

# Teste de Breusch-Pagan para heterocedasticidade
bp_test = het_breuschpagan(residuals, X)
bp_result = pd.Series(bp_test, index=['LM Statistic', 'LM-Test p-value', 'F-Statistic', 'F-Test p-value'])

# Teste de Breusch-Godfrey para autocorrelação
bg_test = acorr_breusch_godfrey(results)
bg_result = pd.Series(bg_test, index=['LM Statistic', 'LM-Test p-value', 'F-Statistic', 'F-Test p-value'])

# Criando a tabela de teste de resíduos
residuals_test_table = pd.DataFrame({'Breusch-Pagan': bp_result, 'Breusch-Godfrey': bg_result})

residuals_test_table 
```

- Indício de auto correlação, vamos testar o Lag -1.

```python
from statsmodels.tsa.stattools import coint, adfuller

# Teste ADF nos resíduos
adf_test = adfuller(residuals)
adf_result = pd.Series(adf_test[0:2], index=['Test Statistic', 'p-value'])

# Criando a tabela de teste de resíduos
residuals_test_table = pd.DataFrame({'ADF': adf_result})
residuals_test_table
```

- O modelo é Estacionário na presente especificação, então temos indícios que as variáveis tem correlação de longo prazo.

```python
robust_results= results.get_robustcov_results(cov_type='HAC',
                                 use_t=None,maxlags =1)
    
# Mostre um resumo dos resultados
print(robust_results.summary())
```

- Utilizando a matriz HAC conseguimos corrigir problemas de heterocedasticidade e autocorrelação nos resíduos. No entanto, é importante notar que a matriz HAC apenas corrige os problemas de heterocedasticidade e autocorrelação nos resíduos em termos de inferência estatística, ou seja, ela não corrige esses problemas em termos de previsão. Sendo assim, é necessário utilizar modelos do tipo ARIMA ou SARIMA para fazer previsões precisas.
